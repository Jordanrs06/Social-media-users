```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()
```

```python
sns.set(style="whitegrid")
```

```python
df.columns
```

```python
avg_time_spent = df.groupby('platform')['total_time_spent'].mean()
```

```python
plt.figure(figsize=(10, 6))
avg_time_spent.plot(kind='bar', color='red')
plt.title('Average Time Spent on Each Platform')
plt.xlabel('Platform')
plt.ylabel('Average Time Spent (minutes)')
plt.show()
```

```python
reason_counts = df['reason_for_watching'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(reason_counts, labels=reason_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Reasons for Watching Videos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['total_time_spent'], bins=20, kde=True, color='purple')
plt.title('Distribution of Total Time Spent on Social Media')
plt.xlabel('Total Time Spent (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='total_time_spent', y='addiction_level', hue='gender', palette='coolwarm')
plt.title('Time Spent vs. Addiction Level')
plt.xlabel('Total Time Spent (Minutes)')
plt.ylabel('Addiction Level')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
correlation_matrix = df[['total_time_spent', 'addiction_level', 'satisfaction']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Time Spent, Addiction Level, and Satisfaction')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_time_spent', y='satisfaction', data=df, hue='platform', palette="viridis", s=100)
plt.title('Total Time Spent vs. Satisfaction Level')
plt.xlabel('Total Time Spent (minutes)')
plt.ylabel('Satisfaction Level')
plt.show()
```

```python
platform_gender = df.groupby(['platform', 'gender']).size().unstack()
```

```python
platform_gender.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
plt.title('Platform Usage by Gender')
plt.xlabel('Platform')
plt.ylabel('Number of Users')
plt.legend(title='Gender')
plt.show()
```

```python
age_satisfaction = df.groupby('age')['satisfaction'].mean()
```

```python
plt.figure(figsize=(10, 6))
age_satisfaction.plot(kind='line', marker='o', color='teal')
plt.title('Average Satisfaction Levels Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Average Satisfaction')
plt.grid()
plt.show()
```

```python
numerical_cols = ['total_time_spent', 'addiction_level', 'satisfaction']
sns.pairplot(df[numerical_cols], diag_kind='kde', plot_kws={'alpha': 0.7})
plt.suptitle('Relationships Between Numerical Features', y=1.02)
plt.show()
```
